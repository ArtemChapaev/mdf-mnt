```python
urls = [
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/1',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/2',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/3',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/4',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/5',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/6',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/7',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/8',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/9',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/10',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/11',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/12',
    'https://wikisource.org/wiki/Колхозонь_эряф/1936/06/ССРСть_Конституцияц/13',
]
```

```python
downloaded_pages = {}
```

```python
for url in urls:
    html = requests.get(url).text
    soup = BeautifulSoup(html)
    content = soup.find('div', {'id': 'mw-content-text'})


    downloaded_pages[unquote_plus(url)] = content.text
```

```python
downloaded_pages
```

```python
processed_pages = {}
```

```python
for i, (url, text) in enumerate(downloaded_pages.items()):
    cleaned_text = text.replace("\u2009", " ")
    cleaned_text = cleaned_text.replace("\xa0", " ")
    processed_pages[url] = cleaned_text
```

```python
processed_pages
```

```python
import json

with open('../text_for_align/konstitution.json', 'w') as f:
    json.dump(processed_pages, f, ensure_ascii=False, indent=1)
```
